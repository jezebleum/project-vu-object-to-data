```python
import csv
import pandas as pd
import glob
import html
import re
from tqdm import tqdm_notebook as tqdm
```

### Made the full merged_christmas file through this piece of code (don't run it constantly, otherwise it just adds to the csv file, if necessary delete the csv file and rerun this code).

### now off to alter the review text (html.unescape) and add it to a dataframe

```python
df = pd.read_csv('merged_christmas.csv', sep='\t', converters={'review_text':html.unescape, 'reviewer_name':html.unescape, 'summary':html.unescape})

df=df[~df['rating'].isin(['rating'])]
#df=df[~df['rating'].isin(['1','2','3','4','5','rating'])]
df
```

### let's do some calculations

```python
import numpy as np
from scipy.stats import f_oneway

actually_Xmas = []
not_Xmas = []
xmas_in_text=[]
no_xmas_in_text=[]

for index, row in tqdm(df.iterrows()):
    
    if re.match('[0-9]+-12-(2[4-9]|30)', row['review_date']):
        actually_Xmas.append(float(row['rating']))
        if "christmas" in row['review_text'].lower():
            xmas_in_text.append(float(row['rating']))
        else:
            no_xmas_in_text.append(float(row['rating']))
    else:
        not_Xmas.append(float(row['rating']))
            
print("actually_Xmas:", round(sum(actually_Xmas)/float(len(actually_Xmas)), 2)," vs not_Xmas:", round(sum(not_Xmas) / float(len(not_Xmas)), 2))
print("number of ratings:", len(actually_Xmas),len(not_Xmas))

print(f_oneway(np.array(actually_Xmas),np.array(not_Xmas)))

print("xmas in text:", sum(xmas_in_text)/len(xmas_in_text), "vs no xmas in text", sum(no_xmas_in_text)/len(no_xmas_in_text))
print("number of ratings:", len(xmas_in_text),len(no_xmas_in_text))
print(f_oneway(np.array(xmas_in_text),np.array(no_xmas_in_text)))
```
