### now off to alter the review text (html.unescape) and add it to a dataframe

```python
df = pd.read_csv('other_datasets/others.csv', sep='\t', converters={'review_text':html.unescape, 'reviewer_name':html.unescape, 'summary':html.unescape})

df=df[~df['rating'].isin(['rating'])]
#df=df[~df['rating'].isin(['1','2','3','4','5','rating'])]
df
```

### let's do some calculations

```python
import numpy as np
from scipy.stats import f_oneway

actually_december = []
not_december = []

for index, row in tqdm(df.iterrows()):
    
    if re.match('[0-9]+-12-(2[4-9]|30)', row['review_date']):
        actually_december.append(float(row['rating']))
    else:
        not_december.append(float(row['rating']))
            
print("actually_december:", round(sum(actually_december)/float(len(actually_december)), 2)," vs not_december:", round(sum(not_december) / float(len(not_december)), 2))
print("number of ratings:", len(actually_december),len(not_december))

print(f_oneway(np.array(actually_december),np.array(not_december)))
```
